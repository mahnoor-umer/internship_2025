Assignment 1: Array Creation and Manipulation

1.Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.

2.Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

```python
import numpy as np

## Task 1
## Create a (5,5) array filled with integers between 1 and 20
arr1 = np.random.randint(1, 21, (5, 5))
print("Original 5x5 array:")
print(arr1)

## Replace all the elements in the third column with 1
arr1[:, 2] = 1
print("\nArray after replacing third column with 1:")
print(arr1)

## Task 2
## Create (4, 4) array with values from 1 to 16
arr2 = np.arange(1, 17).reshape(4, 4)
print("\nOriginal 4x4 array:")
print(arr2)

## Replace the diagonal elements with 0
arr2[np.arange(4), np.arange(4)] = 0
print("\nArray after replacing diagonal elements with 0:")
print(arr2)
```

Assignment 9: Structured Arrays

1.Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

2.Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
## Task 1
## Define a structured datatype
dtype1= np.dtype([('name', 'S10'), ('age', 'i4'), ('weight', 'f4')])

## Create a structured array
arr1 = np.array([('John', 25, 70.5),
                  ('Maria', 31, 60.0),
                  ('Bobz', 42, 80.0)], dtype=dtype1)

## Sort the age
sorted_age= np.sort(arr1['age'])

## Print the sorted array
print( "Sorted_age:\n", sorted_age)

## Task 2
## Define a structured datatype with fields x and y
dtype2 = np.dtype([('x', 'i4'), ('y', 'i4')])

## Create a structured array
arr2 = np.array([(1, 2),
                  (3, 4), 
                  (5, 6)], dtype=dtype2)

## Extract the x and y values
x_values = arr2['x'].astype(float)
y_values = arr2['y'].astype(float)

## Number of points
n = len(arr2)

# Compute pairwise Euclidean distances
print("Pairwise Euclidean Distances:")
for i in range(n):
    for j in range(i+1, n):
        dist = np.sqrt((x_values[i] - x_values[j])**2 + (y_values[i] - y_values[j])**2)
        print(f"Distance between point {i} and {j}: {dist:.2f}")
```

Assignment 10: Masked Arrays

1.Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

2.Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
import numpy as np
import numpy.ma as ma
## Task 1
## Create an array of shape (4,4) with random integers
arr1 = np.random.randint( 1, 30, (4, 4))

## Mask the elements greater than 10
masked1 = ma.masked_greater(arr1, 10)

## Compute the sum of unmasked elements
sum1 = np.sum(masked1)

## Print the results
print( "Original array: \n", arr1 )
print( "Masked array: \n", masked1 )
print( "Sum of unmasked elements: ", sum1 )

## Task 2
## Create an array of shape (3,3) with random integers
arr2 = np.random.randint( 1, 30, (3, 3))
print("\nOriginal array: \n", arr2 )

## Mask the diagonal elements
mask_diagonal = np.eye(3, dtype=bool)

## Create masked array
masked2 = ma.array(arr2, mask=mask_diagonal)
print("\nMasked Array (diagonal elements masked):\n", masked2)

## Compute mean of unmasked elements
mean_unmasked = masked2.mean()
print("\nMean of unmasked elements:", mean_unmasked)

## Replace masked (diagonal) values with mean
filled_array = masked2.filled(mean_unmasked)
print("\nFinal Array (diagonal replaced with mean):\n", filled_array)

```
